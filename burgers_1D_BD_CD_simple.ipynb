```python
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns
```

```python
nt = 1001
nx = 1001

nu = 0.1
#nu = (0.01/math.pi)

dt = 0.001
#dx = 2/ (nx-1)

x= np.linspace(-1,1, nx)

dx = abs(x[0]-x[1])
```

```python
u = np.zeros(nx)
un = np.zeros(nx)

uf = np.zeros((nt,nx))
```

```python
u = -np.sin(math.pi * x)
```

```python
len(u)
```

```python
uf[0,:] = u
```

```python
#u
```

```python
for n in range (1, nt):
    un = u.copy()
    for i in range(1, nx-1):
        u[i] = un[i] - un[i] * (dt/dx) *(un[i] - un[i-1]) + (nu * dt / (dx**2)) * (un[i+1]-2*un[i]+un[i-1])
        uf[n,i] = u[i]
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,u, label = "u Solution")
plt.title("final time step sol.")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
plt.plot(uf[0,:])
plt.plot(uf[-1,:])
plt.grid(True)
plt.show()
```
